```python
# Dependencies
import numpy as np
import pandas as pd
from sqlalchemy import create_engine
```

```python
# set path to csv
global_path = ("regional-global-daily-latest.csv")
aus_path = ("regional-au-daily-latest.csv")
brazil_path = ("regional-br-daily-latest.csv")
switz_path = ("regional-ch-daily-latest.csv")
us_path = ("regional-us-daily-latest.csv")
vietnam_path = ("regional-vn-daily-latest.csv")
souaf_path = ("regional-za-daily-latest.csv")
```

```python
# read in csv/display
global_df = pd.read_csv(global_path)
global_df.head()
```

```python
aus_df = pd.read_csv(aus_path)
aus_df.head()
```

```python
brazil_df = pd.read_csv(brazil_path)
brazil_df.head()
```

```python
switz_df = pd.read_csv(switz_path)
switz_df.head()
```

```python
us_df = pd.read_csv(us_path)
us_df.head()
```

```python
vietnam_df = pd.read_csv(vietnam_path)
vietnam_df.head()
```

```python
souaf_df = pd.read_csv(souaf_path)
souaf_df.head()
```

```python
# merge datasets
merged_df = global_df.append([souaf_df, vietnam_df, us_df, switz_df, brazil_df, aus_df])
merged_df.head()
```

```python
# save to csv
merged_df.to_csv("merged_df.csv", encoding="utf-8", index=False)
```
